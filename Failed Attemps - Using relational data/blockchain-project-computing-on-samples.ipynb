```python
col_names = ["Transaction_Hash", "HashFirstInputTransaction","IndexOfFirstInput","HashSecondInputTransaction","IndexOfSecondInput"]
df=pd.read_csv('/kaggle/input/2015-bitcoin-data/inputs2015_1.txt',sep='\t',names=col_names)
df=df.reset_index()
df.rename(columns={"index": "Timestamp"})
```

```python
output_col_names = ["Transaction_Hash", "Hash_First_Output_Transaction","Amount_First_Output","HashSecondOutputTransaction","Amount_Second_Output"]
df_output=pd.read_csv('/kaggle/input/2015-bitcoin-data/outputs2015_1.txt',sep='\t',names=output_col_names)
df_output=df_output.reset_index()
df_output.rename(columns={"Col1": "Timestamp"})
```

```python
final_df=df.merge(df_output, on='Transaction_Hash', how='left')  
```
